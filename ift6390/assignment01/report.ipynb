```python
import numpy as np
import math 
%pylab inline
import matplotlib.pyplot as plt

class diagonal_gaussian_parametric:
    """
    1. Implement a diagonal Gaussian parametric density estimator. It 
    will have to work for data of arbitrary dimension d. As seen in 
    the labs, it should have a train() method to learn the parameters 
    and a method predict() which calculates the log density.
    """
    def __init__(self):
        pass
    
    def train(self, train_data):
        
        # if only one test_data is passed, add dummy dimension
        if len(np.shape(train_data)) == 1:
            train_data = np.expand_dims(train_data, axis=1)

        self.train_data = train_data
        
        try:
            self.n, self.d = np.shape(self.train_data)
        except:
            self.n = np.shape(self.train_data)
            self.d = 1

        self.mu = np.sum(self.train_data, axis=0) / self.n
        self.sigma = np.cov(self.train_data.T) * np.eye(self.d)


    def predict(self, test_data, nll=True):

        # if only one test_data is passed, add dummy dimension
        if len(np.shape(test_data)) == 1:
            test_data = np.expand_dims(test_data, axis=1).T

        self.test_data = test_data
        n_inputs = np.shape(test_data)[0]
        densities = np.zeros(n_inputs)
        
        sigma_inv = np.linalg.inv(self.sigma)
        sigma_det = np.linalg.det(self.sigma)
            
        normalizer = 1 / ((2* np.pi)**(self.d / 2) * np.sqrt(sigma_det))       

        # we treat each input test_data independently
        for i in range(n_inputs):
            
            exponent = (-0.5) * (self.test_data[i, :] - self.mu).T.dot(sigma_inv).dot(self.test_data[i, :] - self.mu)

            if nll:
                p = -np.log(normalizer)*exponent # instead of normalizer*np.exp(exponent)
            else:
                p = normalizer*np.exp(exponent)
            
            # handle edge case where p(x)=0
            if p == 0:
                p = np.finfo(float).eps
                
            densities[i] = p
        
        return(densities)

    
class parzen_density_estimator:
    """
    2. Implement a Parzen density estimator with an isotropic Gaussian 
    kernel. It will have to work for data of arbitrary dimension d. 
    Likewise it should have a train() method and a predict() method 
    that computes the log density.
    """
    def __init__(self):
        pass
    
    def train(self, train_data, sigma=0):
        self.train_data = np.expand_dims(train_data, axis=1)
        
        try:
            self.n, self.d = np.shape(self.train_data)
        except:
            self.n = np.shape(self.train_data)
        
        if sigma == 0:
            self.sigma = np.std(self.train_data) # std because isotropic Gaussian
        else:
            self.sigma = sigma
            
    
    def predict(self, test_data, nll=True):
        
        # if only one test_data is passed, add dummy dimension
        if len(np.shape(test_data)) == 1:
            test_data = np.expand_dims(test_data, axis=1).T

        self.test_data = test_data
        n_inputs = np.shape(test_data)[0]
        n_train = np.shape(self.train_data)[0]
        densities = np.zeros(n_inputs)
        
        normalizer = 1 / ((2*np.pi*self.sigma)**(self.d / 2))
        
        for i in range(n_inputs): 
            
            # calculate average distance between this training point and all test points
            p = 0 
            for j in range(n_train):
            
                # we're using euclidean distanc
                distance = np.sum((self.test_data[i, :] - self.train_data[j, :])**2, axis=0)
                exponent = (-0.5) * (distance**2 / self.sigma)
                
                if nll:
                    p += -np.log(normalizer)*exponent # normalizer*np.exp(exponent)
                else:
                    p += normalizer*np.exp(exponent)
            
            # handle edge case where p(x)=0
            if p == 0:
                p = np.finfo(float).eps
                
            # save the average kernel values across all training points
            densities[i] = p/n_inputs
        
        return(densities)

```

```python
# test models
iris = np.loadtxt("iris.txt")
test_data_1 = np.array([1,2,3,4,5])
test_data_2 = np.array([[1,2,3,4,5], [2,30,40,50,60]])

test = diagonal_gaussian_parametric()
test.train(iris)

print('parametric 1 input : {}'.format(test.predict(test_data_1)))
print('parametric 2 inputs: {}'.format(test.predict(test_data_2)))

test = parzen_density_estimator()
test.train(iris)

print('nonparametric 1 input : {}'.format(test.predict(test_data_1)))
print('nonparametric 2 inputs: {}'.format(test.predict(test_data_2)))
```

# 1D Densities

From the Iris dataset examples, choose a subset corresponding
to one of the classes (of your choice), and one of the characteristic
features, so that we will be in dimension d = 1 and produce a
single graph (using the plot function) including:

+ the data points of the subset (displayed on the x axis).
+ a plot of the density estimated by your parametric Gaussian estimator.
+ a plot of the density estimated by the Parzen estimator with a
hyper-parameter Ïƒ (standard deviation) too small.
+ a plot of the density estimated by the Parzen estimator with the
hyper-parameter Ïƒ being a little too big.
+ a plot of the density estimated by the Parzen estimator with the
hyper-parameter Ïƒ that you consider more appropriate. Use a
different color for each plot, and provide your graph with a clear
legend.
+ Explain how you chose your hyper-parameter Ïƒ.

```python
# 1d plotting tools

def make_density(x, mdl):
    density = np.zeros(len(x))
    for i, val in enumerate(x):
        output = mdl.predict(np.atleast_2d(val), nll=False)
        density[i] =output

    return(density)
```

```python
# load data
iris = np.loadtxt("iris.txt")
idx = np.where(iris[:,4] == 1)
iris_subset = np.squeeze(iris[idx, 0:1])
n = len(iris_subset)

# general plotting variables
n_bins = 100
axes_min = np.min(iris_subset)-1
axes_max = np.max(iris_subset)+1
alpha=0.75

small_sig = 0.0001
large_sig = 1.0
moyen_sig = 0.1

x = np.atleast_2d(np.linspace(axes_min, axes_max, n_bins)).T

# (a) -- raw data
plt.figure(figsize=(12, 8))
plt.scatter(iris_subset, np.zeros(len(iris_subset)))

# (b) -- parametric model
flower = diagonal_gaussian_parametric()
flower.train(iris_subset)

density = flower.predict(x, nll=False)
plt.plot(x, density, color='black', alpha=alpha)

# (c) -- nonparametric small sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=small_sig)
density = flower.predict(x, nll=False)
plt.plot(x, density, color='blue', alpha=alpha)

# (d) -- nonparametric large sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=large_sig)
density = flower.predict(x, nll=False)
plt.plot(x, density, color='red', alpha=alpha)

# (e) -- nonparametric just-right sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=moyen_sig)
density = flower.predict(x, nll=False)
plt.plot(x, density, color='green', alpha=alpha)

# legend
plt.legend(['parametric', 
            'nonparanetric small sigma',
            'nonparametric large sigma',
            'nonparametric just-right sigma',
            'raw data'])

plt.title('1d estimations')
plt.xlabel('feature 1')
plt.ylabel('p(x)')
plt.show()
```

We chose sigma based on how it looked on the plot. The very large sigma (1.0) was chosen to encompase lots of empty feature space and the very small sigma (0.01) was chosen to tightly wrap each data point. For the 'just right' sigma (0.1), we chose a value that covered the data distribution entirely without it encompasing too much empty feature space.

# 2D densities
Now add a second characteristic feature of Iris, in order

```python
# plotting tools
# https://scipython.com/blog/visualizing-the-bivariate-gaussian-distribution/
def multivariate_gaussian(pos, mu, Sigma):
    """Return the multivariate Gaussian distribution on array pos.

    pos is an array constructed by packing the meshed arrays of variables
    x_1, x_2, x_3, ..., x_k into its _last_ dimension.

    """
    n = mu.shape[0]
    Sigma_det = np.linalg.det(Sigma)
    Sigma_inv = np.linalg.inv(Sigma)
    N = np.sqrt((2*np.pi)**n * Sigma_det)
    # This einsum call calculates (x-mu)T.Sigma-1.(x-mu) in a vectorized
    # way across all the input variables.
    fac = np.einsum('...k,kl,...l->...', pos-mu, Sigma_inv, pos-mu)

    return np.exp(-fac / 2) / N


def make_z(axes_min, axes_max, n, mu, sigma):

    # Our 2-dimensional distribution will be over variables X and Y
    X = np.linspace(axes_min, axes_max, n)
    Y = np.linspace(axes_min, axes_max, n)
    X, Y = np.meshgrid(X, Y)

    # Pack X and Y into a single 3-dimensional array
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X
    pos[:, :, 1] = Y

    # The distribution on the variables X, Y packed into pos.
    Z = multivariate_gaussian(pos, mu, sigma)

    return(X, Y, Z)


def sum_densities_2d(axes_min, axes_max, n_bins, n, x, data, sigma):
    full_model = np.zeros((n_bins, n_bins))
    for i in range(n):
        X, Y, Z = make_z(axes_min, axes_max, n_bins, data[i], sigma)
        full_model += Z
    full_model /= n
    
    return(X, Y, full_model)
    
```

```python
# new data subset
iris_subset = np.squeeze(iris[idx, 0:2])

# general plotting variables
n_bins = 100
axes_min = np.min(iris_subset)-1
axes_max = np.max(iris_subset)+1
alpha=0.75

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

# add raw data
ax1.scatter(iris_subset[:, 0], iris_subset[:, 1])
ax2.scatter(iris_subset[:, 0], iris_subset[:, 1])
ax3.scatter(iris_subset[:, 0], iris_subset[:, 1])
ax4.scatter(iris_subset[:, 0], iris_subset[:, 1])

# (b) -- parametric model
flower = diagonal_gaussian_parametric()
flower.train(iris_subset)
x, y, z = make_z(axes_min, axes_max, n_bins, flower.mu, flower.sigma)
ax1.contour(x, y, z, colors='red')
ax1.set_title('parametric model sigma={}'.format(flower.sigma[0]))
ax1.set_ylabel('feature 2')

# (c) -- nonparametric model, small sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=small_sig)
x, y, z = sum_densities_2d(axes_min, axes_max, n_bins, n, x, iris_subset, np.eye(2)*flower.sigma)
ax2.contour(x, y, z, colors='red', alpha=alpha)
ax2.set_title('nonparametric mode small sigma={}'.format(small_sig))

# (d) -- nonparametric model, large sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=large_sig)
x, y, z = sum_densities_2d(axes_min, axes_max, n_bins, n, x, iris_subset, np.eye(2)*flower.sigma)
ax3.contour(x, y, z, colors='red', alpha=alpha)
ax3.set_title('nonparametric model large sigma={}'.format(large_sig))
ax3.set_xlabel('feature 1')
ax3.set_ylabel('feature 2')


# (e) -- nonparametric model, just-right sigma
flower = parzen_density_estimator()
flower.train(iris_subset, sigma=moyen_sig)
x, y, z = sum_densities_2d(axes_min, axes_max, n_bins, n, x, iris_subset, np.eye(2)*flower.sigma)
ax4.contour(x, y, z, colors='red', alpha=alpha)
ax4.set_title('nonparametric model just-right sigma={}'.format(moyen_sig))
ax4.set_xlabel('feature 1')

plt.suptitle('2d estimations')
plt.tight_layout()
plt.show()

```

For the 'just right' sigma (0.1), we chose a value that covered the data distribution entirely without it encompasing too much empty feature space.
